# Image Classification

```python
# Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import online data
from sklearn.datasets import load_digits

digits = load_digits()
```

```python
# input variables or input (X)

digits.data.shape
X= digits.data
# Means 1797 pictures and size = 8x8 =64
```

```python
# output lables (y)
digits.target.shape
y= digits.target
```

```python
# Ploting

plt.figure(figsize= (20,4))
for index, (image, label) in enumerate(zip(X[0:10], y[0:10])):
    plt.subplot(1, 10, index+1)
    plt.imshow(np.reshape(image, (8,8)), cmap= plt.cm.gray)
    plt.title('Training: %i\n' % label, fontsize =20)


```

```python
# Ploting

plt.figure(figsize= (20,4))
for index, (image, label) in enumerate(zip(X[0:5], y[0:5])):
    plt.subplot(1, 5, index+1)
    plt.imshow(np.reshape(image, (8,8)), cmap= plt.cm.gray)
    plt.title('Training: %i\n' % label, fontsize =20)
```

```python
# Ploting

plt.figure(figsize= (20,4))
for index, (image, label) in enumerate(zip(X[0:20], y[0:20])):
    plt.subplot(1, 20, index+1)
    plt.imshow(np.reshape(image, (8,8)), cmap= plt.cm.gray)
    plt.title(label, fontsize =20)
```

```python
# help(plt)
```

```python
# split the data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=0)
```

```python
print('Train input data:', X_train.shape)
print('Test input data:', X_test.shape)
print('Train output data:', y_train.shape)
print('Test output data:', y_test.shape)
```

```python
# Train model
from sklearn.linear_model import LogisticRegression
model= LogisticRegression().fit(X_train, y_train)
model
```

```python
# prediction
model.predict(X_test[0:10])
prediction = model.predict(X_test)
prediction
```

```python
# Accuracy test
score= model.score(X_test, y_test)
print('The accuracy score is = ', score)
```

```python
# Confusion matrix
from sklearn import metrics
cm = metrics.confusion_matrix(y_test, prediction)
cm
```

```python
import seaborn as sns
plt.figure(figsize= (9,9))
sns.heatmap(cm, annot=True, fmt= '.3f', linewidths=.5, square= True, cmap= 'Spectral')
plt.ylabel('Actual Output')
plt.xlabel('Predicted Output')
all_sample_title = 'Accuracy Score: {0}'.format(score)
plt.title(all_sample_title, size= 15)
```

```python
# Getting miss classified lables

import numpy as np
import matplotlib.pyplot as plt
index =0 
misclassifiedIndexes = []
for label, predict in zip(y_test, prediction):
    if label != predict:
        misclassifiedIndexes.append(index)
        index+=1
```

```python
# plotting misclassified labels with known labelsv
plt.figure(figsize = (20, 4))
for plotIndex, badIndex in enumerate(misclassifiedIndexes[5:10]):
    plt.subplot(1, 5, plotIndex + 1)
    plt.imshow(np.reshape(X_test[badIndex], (8,8)), cmap= plt.cm.gray)
    plt.title("Predicted:{}, Actual:{}".format(prediction[badIndex], y_test[badIndex]), fontsize=15)
```
